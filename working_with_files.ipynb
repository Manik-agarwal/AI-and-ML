```python
df = pd.DataFrame()

for i in range(1, 429):
    response = requests.get(f"https://api.themoviedb.org/3/tv/top_rated?api_key=b20f3ca4fa2597a7bb72fef875098a4a&language=en-US&page={i}")
    data = response.json()
    
    # Check if 'results' key is in the response and if it's not empty
    if 'results' in data and len(data['results']) > 0:
        results = data['results']
        # Check if all required columns are present
        if all(key in results[0] for key in ['id', 'name', 'overview', 'first_air_date', 'popularity', 'vote_average', 'vote_count']):
            temp_df = pd.DataFrame(results)[['id', 'name', 'overview', 'first_air_date', 'popularity', 'vote_average', 'vote_count']]
            df = pd.concat([df, temp_df], ignore_index=True)
        else:
            print(f"Missing columns in page {i}")
    else:
        print(f"No results found on page {i}")

# Optionally, you can save the DataFrame to a CSV file
# df.to_csv('top_rated_tv_shows.csv', index=False)
```
