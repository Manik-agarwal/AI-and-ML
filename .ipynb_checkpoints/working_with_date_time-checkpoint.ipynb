EXTRACT YEAR MONTH DAYS

```python
date['date_year']=date['date'].dt.year
```

```python
date['date_months']=date['date'].dt.month
```

```python
date['date_day']=date['date'].dt.day
```

```python
date.head()
```

```python
date['date_dow']=date['date'].dt.day_of_week
```

```python
date['day_name']=date['date'].dt.day_name()
```

```python
date.head()
```

```python
date['day_is_weekend']=np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.drop(columns=['product_id','city_id','orders'])
```
